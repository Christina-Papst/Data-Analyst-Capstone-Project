## Extract information from the given web site

You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here
from bs4 import BeautifulSoup
import requests
```

Download the webpage at the url


```python
#your code goes here
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
data  = requests.get(url).text 
```

Create a soup object


```python
#your code goes here
soup = BeautifulSoup(data,"html5lib")
```

Scrape the `Language name`, `Created By` `annual average salary`and `Learning Difficulty`.


```python
#your code goes here
table = soup.find('table')
```

Create a _dataframe_ for scrapped data


```python
#your code goes here
import pandas as pd
```

```python
tables = soup.find_all('table')
```

```python
for index,table in enumerate(tables):
    table_index = index
print(table_index)
```

```python
popular_languages = pd.DataFrame(columns=["Languages", "Created by", "Average annual salary", "Learning Difficulty"])

for row in tables[table_index].tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        languages = col[1].text
        created = col[2].text.strip()
        salary = col[3].text.strip()
        difficulty = col[4].text.strip()
        popular_languages = popular_languages._append({"Languages":languages, "Created by":created, "Average annual salary":salary, "Learning Difficulty":difficulty}, ignore_index=True)

```

```python
popular_languages
```

Save the scrapped data into a file named _popular-languages.csv_


```python
# your code goes here
popular_languages.to_csv("popular-languages.csv", index=False)
```
